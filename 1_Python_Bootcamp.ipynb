## List & Looping statement

Data can be processed using different data structures.

- Dataframe (like Excel) using Pandas
- **List**
- ...

Each value is assigned into an **index** or a **position** ... this is like **address** ... **the first index is zero; the second index is one**

Lists can be thought of as a series of boxes. Each box has a different value. For example, the boxes created by demolist = ['life', 42, 'the universe', 6, 'and', 9] would look like this:

<img src="images/list.gif">

Each box is referenced by its number so the statement demolist[0] would get 'life', demolist[1] would get 42 and so on up to demolist[5] getting 9.

````python
items = ["life", 42, "the universe", 6, "and", 9]

print(items)  # prints all elements
print(items[0])  # print first element ... each value is assigned into an index or a position ... the first index is zero
print(items[1])  # prints second element ... the second index is one
print(items[0:2])  # print the first two items
print(items[-1])  # print the last item
````

````python
# how many elements in the list called items
len(items)
````

````python
# use for loop to print the item values

for item in items:
    print(item)
````

````python
# create a list of string and number values ... create a unique variable name ("collecteditems")
# print all the values in the list "collecteditems"
# display the last value in that list
collecteditems = ["Book","Pen","12","Notes","44"]
print(collecteditems)
print(collecteditems[-1])

````

````python
# how many elements in the list called collecteditems
len(collecteditems)
````

````python
# use for loop to print the values of the collecteditems list
for i in collecteditems:
    print(i)
````
